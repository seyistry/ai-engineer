```python
# using median to determine the missing value in age and footlength

age_med = df['age'].median()
print(age_med)

footlength_med = df['footlgth'].median()
print(footlength_med)


df['age'] = df['age'].fillna(age_med)
df['footlgth'] = df['age'].fillna(footlength_med)

# checking if there is any missing file after the correction
df.isna().sum()
```

```python
# List of numerical features
numerical_features = ['age', 'hdlngth', 'skullw', 'totlngth', 'taill', 'footlgth', 'earconch', 'eye', 'chest', 'belly']
# Define units for numerical features
units = {
    'age': '',
    'hdlngth':'' ,
    'skullw':'' ,
    'totlngth':'' ,
    'taill': '',
    'footlgth':'' ,
    'earconch': '',
    'eye': '',
    'chest': '',
    'belly': ''
}

# Compute mean, median, and mode for each numerical feature using a for loop
for col in numerical_features:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode().iloc[0]
    unit = units[col]
    print(f"\nMeasure of Center for {col} in {unit}:")
    print(f"  Mean: {mean_val:.2f} {unit}")
    print(f"  Median: {median_val:.2f} {unit}")
    print(f"  Mode: {mode_val:.2f} {unit}")
```
