**Why Should We Preprocess Data Before Building The ML Model?**

**NOTE**: *I will introduce some new terms here, don't get worried if you dont understand them yet, but is good you are aware of their existence. I will bolden then for emphasis*

Lets break it down, the effectiveness of our model is at the mercy of how well our features are engineered, and **feature engineering** is at the mercy of data preprocesing. Data preprocessing is crucial in machine learning because raw data is often incomplete, inconsistent, and noisy.Without proper preprocessing, **ML algorithms** may learn incorrect patterns, leading to poor performance.


So lets answer the question now.
1. To improve data quality: After cleaning data **features** might still contain errors, incosistencies and missing features that can affect the performance of the **model**. So preprocesing prepares the dataset for **feature selection** and **feature extraction** helping.

2. Enhance Model Performance: Most ML algorithms learns better when the features are  **scaled**, **normalized** and structured properly. So proper preprocessing ensures that features contribute meaningfully to **predictions**.

3. Prevent Models From **Overfitting**, **Underfitting**, **High Variance** and **Bias**: Handling **imbalanced dataset**s prevents the model from being biased towards the dominant **class**. So proper preperocessing ensures that the features contributes meaningfully to predictions of the model.

4. Ensures Algorithm Compatibility: Many ML algorithms require numeric inputs (e.g., **encoding** categorical variables).
Some algorithms, like KNN and neural networks, are sensitive to scale, requiring normalization.

5. Improves Model Accuracy And Efficiency: Well preprocessed data helps models learn better patterns, leading to improved **accuracy** and **generalization**.
Reduces **computational complexity** by eliminating **irrelevant features**.

**Things To Note When Preprocessing**

`fit_transform()`method

It is commonly used in data preprocessing in machine learning, particularly when using tools like Scikit-learn for tasks such as imputation, scaling, normalization, or encoding data. **Methods for Handling Outliers**

To know which of the methods to use, consider domain knowledge, visualize the data to understand the distribution and experiment. Below are method for handling outliers.
1. Removal
2. Capping(Winsorizing or Clipping)

**Outlier Removal**:
To use removal you must have a certain level of knowledge on the field that that data belongs to. If you are not knowledgable about the data, you should at least consult with an expert in that field or speak with the business organization that provided the data. **Capping**: From our example in data cleaning, the method I used to handle outliers there is the capping method. This method is used when outliers truly represents valid data points but you want to reduce their influences on the model you are building. It preserves more information compared to the removal method.


In our illustration, we will be using Z-score

**How To Manually Calculate Z-score**

This is the formula here.

`Z = {̱X - μ}÷{σ}`

X ⟶ Each data point (observation) in the dataset or selected column

μ ⟶ The mean (average) of the dataset or selected column

σ ⟶ The standard deviation of the dataset column

**How Does Z-Score Work?** _its ok to know it, right?_



Z-score **standardizes** the data by measuring how far X is from the **mean** in terms of **standard deviations**.

When you are done calculating, look through all the Z-score to see anything unusual.

* A Z-score of 0 indicates the value is exactly at the mean.

* A Z-score of +1 indicates the value is one standard deviation above the mean.

* A Z-score of -1 indicates the value is one standard deviation below the mean.

* Any score above 3 and below -3 is considered an outlier, that is, X is an outlier if Z>3 or Z<−3


**Please Note:** That the `stats.zscore()` function calculates the Z-scores for all **numerical column** in a DataFrame by default, but you can choose a specific column as it applies to the dataset you are working with.

**Lets practice...**

````python
from scipy import stats # This is a statistics library
z_scores = stats.zscore(features) # Calculate the z-score for the whole dataframe

# lets print the z-scores
print(z_scores)
````

````python
# Lets filter out the outliers --> features[condition]

# abs()-->help to handle -tive value if there are any...

# feature_with_no_outliers = df[abs(stats.zscore(features["sepal length (cm)"])) < 3] ---> This will filter for only one column

# lets create a copy of our dataset

handle_outliers_1 = features.copy()
features_with_removed_outliers = handle_outliers_1[abs((z_scores < 3).all(axis=1))] # .all(axis = 1)is signifying that I am selecting all the columns
````

**Capping Method**

I am using the numpy function `np.where()` to create a condition to help apply the capping


`np.where(condition, [x, y])`

*  The condition: `np.abs(z_scores.iloc[:, i]) > threshold`

* The x:  `handle_outliers_2[col].clip(lower=outliers.min(), upper=outliers.max())`,

* The y: `handle_outliers_2[col]`


The condition first convert to absolute values and select all rows(:) in the current column(i) from the (z_scores) DataFrame and check if its greater than the threshhold

The `clip()` method caps both the lower and upper limit if the condition is `True`. Finally, all columns that capping is applied to is returned.


**Note:** I will comment the code alot for better understanding, once you understood it, kindly remove the comments within the code so with can have a clean code.

````python
import numpy as np

# lets create a copy of our dataframe
handle_outliers_2 = features.copy()

# lets define the threshold first
threshold = 3

# lets apply capping to our calculated z-scores
### Lets apply a little bit of complexity to it incase you are going to be selecting the numeric columns in your own dataset, it works here but doesnt really applies to this case though.
### creating a loop to handle both the limits and the capping of the selected columns

for i, col in enumerate(handle_outliers_2.select_dtypes(include=[np.number]).columns):
  # lets select the outliers
  outliers = handle_outliers_2[col][abs(stats.zscore(handle_outliers_2[col])) > threshold]
  # lets cap anything outside the threshold
  # Use.iloc[] to access the z_scores values
  handle_outliers_2[col] = np.where(np.abs(z_scores.iloc[:, i]) > threshold, handle_outliers_2[col].clip(lower=outliers.min(), upper=outliers.max()), handle_outliers_2[col]) #Using clip to cap outliers within the IQR range
````

#####**5. Dimensionality Reduction**

Have tried listening to someone recounting an event, and you hear the person say, "*to cut the long story short, let me just tell you what happened*".The person is in a way saying, let's just filter out the unimportant details...

When it comes to dimensionality reduction, it has to do with keeping out or filtering out the unimportant features(dimensions) to keep the import onces only, but it does it without loosing out the import information. Hence, will retain the features that will contribute most to the training.

Dimensionality reduction helps;
1. Understand and visualize  the data in a more manageable way.
2. For faster computation during model training.
3. With improved model performance.

**What is the intuition behind dimensionality reduction?**

There are different techniques used, but what they do majorly is finding  *patterns* and *relationships* in your data to create new, "summarized" variables. The number of dimensions are reduced while still capturing the essence of the information.

**Techniques used for Dimensionality Reduction**

1. Linear Discriminant Analysis (LDA): Specifically designed for classification problems, it finds dimensions that best separate different classes.

2. t-Distributed Stochastic Neighbor Embedding (t-SNE): Focuses on preserving the local structure of your data, making it useful for visualization.

3. Principal Component Analysis (PCA): Finds the directions of greatest variance in your data and creates new variables (principal components) along those directions.

````python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
````

````python
# Lets load dataset
digits = load_digits()
````

````python
# Lets view our digit data(target)
fig, axes = plt.subplots(2, 5, figsize=(8, 4))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Digit: {digits.target[i]}")
    ax.axis("off")
plt.show()
````

````python
# The data is in array format, lets convert to pandas dataframe to have a better view
X = digits.data  # Features (64-dimensional)
y = digits.target  # Labels (0-9)
````

````python
# Convert the data to pandas dataframe and series
X = pd.DataFrame(X)
y = pd.Series(y)
````

````python
# lets inpect the data
X.head()
````

````python
X.columns
````

````python
print(f"The rows is: {X.shape[0]}")
print(f"The columns is: {X.shape[1]}")
````

````python
# Lets make a copy of X
X_dimensions = X.copy()

# Create an instance of your PCAC() class and choose the numbers of your principal compoents
# choose the numbers ofyour principal components(n_components)
pc = PCA(n_components=6)

# fit and transform the features to reflect the reduced dimensions
Reduced_X_Dimensions = pc.fit_transform(X_dimensions)
````

````python
Reduced_X_Dimensions.shape
````

````python
print(Reduced_X_Dimensions)
````

Using PCA() our features have been reduced from `64 features`  to `6 feature` and still keeping the relevant information of the dataset.

````python

# LEts plot a scatter plot for our PCA results
# We will plot our first and second PCs  to see their relatedness
plt.figure(figsize=(8,6))
plt.scatter(Reduced_X_Dimensions[:, 0],Reduced_X_Dimensions[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.colorbar(label="Digit Class")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Digits Dataset")
plt.show()
````

**Interpreting PCA ussing the Scatter plot**:

The scatter plot has helped us to display the relationship between the first two principal components (PCs) derived from the original dataset.

So by visualizing the data in this lower-dimensional space, it is easy to can gain insights into patterns and clusters within our data.


*  Points that are close together on the scatter plot are likely to be similar to each other based on the features captured by PC1 and PC2. The distinct clusters of points with different colors, suggests that PCA has been able to separate different digit classes effectively in this lower-dimensional space.

* The spread of the points along each axis reflects the amount of variance captured by that PC. A wider spread along an axis indicates that that PC captures more variance in the data.

* We can observe the relationship between the two PCs by looking at the overall shape of the scatter plot. For example, if the points form a linear pattern, it suggests a strong correlation between the two PCs. In this case there is no linear relationship but we can see more clusters and few spreads. Meaning that PCA has been able to capture the essential features that differentiate between these classes.

#####**6.Encoding Categorical Variables**

In machine learning, **encoding** refers to the process of converting categorical data that is, the non-numericals into a numerical format that can be interpreted by algorithms. Most machine learning algorithms, especially traditional ones (like Logistic Regression, SVM, and Neural Networks), require numerical input to make calculations, so encoding categorical variables is quite important to the model training process.

**The Intuition Behind Encoding**

Categorical variables represent data that can take on a limited number of values, like colors, product categories, or geographical locations. While these variables contain valuable information, most algorithms cannot work directly with such data because they are non-numerical.

So what **Encoding** does is to transforms these categorical values into numerical *representations* that algorithms can understand.

But there is the challenge of **preserving the meaning and relationships** within the data when transforming it into numerical form. For example, we don’t want the encoding of "Red", "Green", "Blue" to imply any inherent numerical relationship like "Red is less than and Green is less than Blue", especially when the variable has no such order. For ordinal variables, we might want to preserve the relationship (e.g., "Low", "Medium", "High" could be encoded as 0, 1, 2).

**Examples of Categorical Features**

1. **Nominal (Unordered) Categorical Features**
   - **Color**: `Red`, `Blue`, `Green`
   - **Country**: `Nigeria`, `Ghana`, `Togo`
   - These values don’t have a meaningful order, so techniques like **One-Hot Encoding** or **Binary Encoding** are often used.

2. **Ordinal (Ordered) Categorical Features**:
   - **Education Level**: `High School`, `Bachelor`, `Master`, `PhD`
   - **Product Rating**: `Low`, `Medium`, `High`
   - These variables have a natural ordering, so methods like **Ordinal Encoding** or **Label Encoding** are suitable because they preserve this order.

3. **High Cardinality Categorical Features**:
   - **City Name**: `Warri`, `Enugu`, `Benin`, `Jos`, `Lagos`, etc.
   - **Product ID**: `P1234`, `P5678`, `P9123`, etc.
   - When there are many distinct categories, methods like **Binary Encoding**, **Frequency Encoding**, or **Hash Encoding** are effective in reducing dimensionality and preventing data explosion.

**Why Is Encoding Important?**

1. Proper encoding makes the relationships between data points more understandable for the model. Many datasets contain a mix of numerical and categorical features. Proper encoding of categorical data allows for seamless integration with numerical features, improving the model's ability to make sense of the whole dataset.
2. Most algorithms can’t process categorical variables directly, so encoding is a critical preprocessing step.

**Table 1**


---



| **Encoding Types**   | **Characteristics** | **Compatible Algorithms** |
|----------------------|---------------------|---------------------------|
| **Label Encoding**    | Assigns unique integer values to categories | Tree-based models (Decision Trees, Random Forest, XGBoost) |
| **Ordinal Encoding**  | Similar to label encoding but allows predefined order | Tree-based models, Linear models |
| **One-Hot Encoding**  | Creates separate binary columns for each category | Linear models (Logistic Regression, SVM), Tree-based models |
| **Binary Encoding**   | Converts categories into binary representation, reducing dimensionality | Tree-based models, Some linear models |
| **Target Encoding**   | Replaces categories with the mean of the target variable | Tree-based models (Random Forest, XGBoost, LightGBM) |
| **Mean Encoding**     | Assigns category values based on the mean of the target | Tree-based models |
| **Frequency Encoding**| Replaces categories with their frequency count in the dataset | Tree-based models |
| **Hash Encoding**     | Uses a hash function to assign a fixed-length vector to each category | Tree-based models, Deep Learning models |
| **Weight of Evidence (WOE) Encoding** | Assigns a weight based on the distribution of the target variable | Logistic Regression, Decision Trees |


**Table 2**


---



| **Encoding Types**   | **Advantages** | **Disadvantages** |
|----------------------|----------------|-------------------|
| **Label Encoding**    | Simple to implement; works well for tree-based models | Introduces artificial ordinal relationships where none exist |
| **Ordinal Encoding**  | Maintains ordinal relationships; works well when order is meaningful | Can mislead models if order is not truly ordinal |
| **One-Hot Encoding**  | Preserves category independence; improves performance in linear models | Can cause high dimensionality when categories are many |
| **Binary Encoding**   | Reduces dimensionality compared to One-Hot Encoding | Can be harder to interpret compared to One-Hot Encoding |
| **Target Encoding**   | Captures information about target distribution; reduces dimensionality | High risk of data leakage; prone to overfitting |
| **Mean Encoding**     | Effective when categories have strong predictive power | Overfitting risk; must be used with cross-validation |
| **Frequency Encoding**| Simple to implement; reduces dimensionality | Assumes frequency correlates with target, which is not always true |
| **Hash Encoding**     | Reduces dimensionality significantly; works well in large datasets | Introduces collisions (different categories can have the same encoding) |
| **Weight of Evidence (WOE) Encoding** | Works well with logistic regression; handles imbalanced datasets effectively | Requires domain knowledge; not suitable for all use cases |

---
